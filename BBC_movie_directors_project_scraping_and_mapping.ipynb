# **Where directors were born**

```python
# put each director info inside a dictionary and them create a list with it; 
# this will be used to create a dataframe later

director_dicts = []
from tqdm.auto import tqdm

for names in tqdm(all_directors):
    try:
        string = names.replace(" ","+").lower()
        href = "https://www.imdb.com/find?q=" + string
        raw_html = requests.get(href).content
        soup_doc = BeautifulSoup(raw_html, "html.parser")
        table_result = soup_doc.table
        td_result = table_result.find(class_="result_text")
        director_info = {}
        director_info["name"] = names
        director_info["link"] = td_result.a["href"]
        director_dicts.append(director_info)
    except:
        director_info = {}
        director_info["name"] = names
        director_info["link"] = "failed"
        director_dicts.append(director_info)
        
director_dicts
```

```python
#open the link inside the dictionary and their birthplace

for dicts in tqdm(director_dicts):
    try:
        print(dicts)
        direct = 'https://www.imdb.com'
        direct_link = direct+dicts['link']
        raw_html = requests.get(direct_link).content
        soup_doc = BeautifulSoup(raw_html, "html.parser")
        origin = soup_doc.find(id='name-born-info')('a')[2]
        origin_text = origin.text
        print(origin_text)
        dicts['birth_place'] = origin_text
       
    except:
        print('-------')
        print(f'i did not find anything for{director_dicts}')
        print('-------')
        dicts['birth_place'] = 'empty'

director_dicts
```

```python
#reorganize to become geojson file

def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```
